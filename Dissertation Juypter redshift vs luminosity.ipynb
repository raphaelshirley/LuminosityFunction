```python
from astropy.table import Table
import matplotlib.pyplot as plt
import numpy as np
t = Table.read('ELAIS-N1_20171020-test.fits')['redshift','cigale_dustlumin']

x = []
y = []

for i in t['redshift']:
    x.append(i)
    
for i in t['cigale_dustlumin']:
    y.append(i)

plt.scatter(x,np.log(y), s = 10)
plt.title('Herschel Satellite data')
plt.xlabel('redshift')
plt.ylabel('log(luminosity)')
plt.grid()
plt.show()
```
